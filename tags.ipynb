# Extract tags and select top 10 tags

```python
questions = so.questions(df)
tag_df = so.extract_tags(questions)
```

### Top 10 tags

```python
tag_counts = tag_counts[:10]
```

### Plot top 10 tags

```python
sns.barplot(data=tag_counts)
plt.xlabel('Tag')
plt.ylabel('Frequency')
plt.gca().tick_params(axis='x', rotation=45)
```

### Questions with top 10 tags

```python
tag_df_top10 = tag_df[tag_df['parsed_tags'].isin(tag_counts.index)]
```

```python
tag_df_top10
```

### Control and treated questions with top 10 tags

```python
tag_T_df = tag_df[tag_df['T'] == 1]
tag_C_df = tag_df[tag_df['T'] == 0]
tag_T_counts = tag_T_df['parsed_tags'].value_counts()
tag_C_counts = tag_C_df['parsed_tags'].value_counts()
```

#### Top 10 tags after the treatment

```python
tag_T_counts[:10]
```

#### Top 10 tags for control

```python
tag_C_counts[:10]
```

#### Plot top 10 tags in treated

```python
sns.barplot(x='parsed_tags', y='count', data=pd.DataFrame(tag_T_counts[:10]), zorder=3, color=sns.color_palette()[0])
plt.title('Tag Frequency (22/23)')
plt.xlabel('Tag')
plt.ylabel('Frequency')
plt.gca().tick_params(axis='x', rotation=45)
plt.grid(color='lightgray', linestyle='--', linewidth=0.5, zorder=0)
plt.tight_layout()

filepath = Path('figures/tags/tt.pdf')
filepath.parent.mkdir(parents=True, exist_ok=True)
plt.savefig(filepath)
```

#### Plot top 10 tags in control

```python
sns.barplot(x='parsed_tags', y='count', data=pd.DataFrame(tag_C_counts[:10]), zorder=3, color=sns.color_palette()[1])
plt.title('Tag Frequency (21/22)')
plt.xlabel('Tag')
plt.ylabel('Frequency')
plt.gca().tick_params(axis='x', rotation=45)
plt.grid(color='lightgray', linestyle='--', linewidth=0.5, zorder=0)
plt.tight_layout()

filepath = Path('figures/tags/tc.pdf')
filepath.parent.mkdir(parents=True, exist_ok=True)
plt.savefig(filepath)
```

#### Plot top 10 overall and how they changed after the treatment

```python
tag_T_df_top10 = tag_df_top10[tag_df_top10['T'] == 1]
tag_C_df_top10 = tag_df_top10[tag_df_top10['T'] == 0]
tag_T_counts = tag_T_df_top10['parsed_tags'].value_counts()
tag_C_counts = tag_C_df_top10['parsed_tags'].value_counts()
tags_combined = pd.concat([pd.DataFrame(tag_T_counts).assign(Period='22/23'), pd.DataFrame(tag_C_counts).assign(Period='21/22')])
```

```python
sns.barplot(x='parsed_tags', y='count', hue='Period', data=tags_combined, zorder=3)
plt.title('Tag Frequency')
plt.xlabel('Tag')
plt.ylabel('Frequency')
plt.gca().tick_params(axis='x', rotation=45)
plt.grid(color='lightgray', linestyle='--', linewidth=0.5, zorder=0)
plt.tight_layout()

filepath = Path('figures/tags/ttc.pdf')
filepath.parent.mkdir(parents=True, exist_ok=True)
plt.savefig(filepath)
```

# Tags time series (Top 10 tags)

### Question volume per tag

```python
for tag in tag_counts.index:
    tag_T_resampled, tag_C_resampled = so.ts(tag_df_top10[tag_df_top10['parsed_tags'] == tag])
    treated = tag_T_resampled[['parsed_tags']].count()
    control = tag_C_resampled[['parsed_tags']].count()
    escaped_tag = tag.replace('#', r'\#')
    so.plot_ts(treated, control, 'parsed_tags', fr'Questions with $\mathit{{{escaped_tag}}}$ tag', f'tags/{tag}.pdf')
```
