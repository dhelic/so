```python
import so
import gc
import numpy as np
```

## Load the data

```python
df = so.load_data()
```

```python
df.info()
```

### Log transform of post lengths

```python
df['log_lc'] = np.log(df['line_count'])
```

### Extract questions and answers

```python
questions = so.questions(df)
answers = so.answers(df)
```

## Compute the distributions

### Question length

```python
counts = questions['line_count'].value_counts(bins=50, sort=False)
print('Question Length')
print(counts)
```

```python
counts = questions['log_lc'].value_counts(bins=50, sort=False)
print('Question Log Length')
print(counts)
```

```python
filenames = ['android.csv', 'csharp.csv', 'java.csv', 'other.csv', 'python.csv', 'r.csv', 'web.csv']
names = ['android', 'c#', 'java', 'other', 'python', 'r', 'web']
```

```python
for i, (name, filename) in enumerate(zip(names, filenames)):
    print(f'Processing: {name}')
    
    escaped_name = name.replace('#', r'\#')
    
    df = so.load_data(filename=f'data/csv/tags/{filename}')
    df.info()

    df['log_lc'] = np.log(df['line_count'])
    df['loc'] = df['code'].str.count('\n') + 1
    df['log_loc'] = np.log(df['loc'])
    
    # question length
    counts = df['line_count'].value_counts(bins=50, sort=False)
    print('Question Length')
    print(counts)

    # question log length
    counts = df['log_lc'].value_counts(bins=50, sort=False)
    print('Question Log Length')
    print(counts)

    # question views
    counts = df['ViewCount'].value_counts(bins=50, sort=False)
    print('Question Views')
    print(counts)

    # question score
    counts = df['Score'].value_counts(bins=50, sort=False)
    print('Question Score')
    print(counts)

    # loc
    counts = df['loc'].value_counts(bins=50, sort=False)
    print('LOC')
    print(counts)

    # log loc
    counts = df['log_loc'].value_counts(bins=50, sort=False)
    print('Log LOC')
    print(counts)
```
