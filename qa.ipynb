```python
questions_T_resampled, questions_C_resampled = so.ts(questions)
answers_T_resampled, answers_C_resampled = so.ts(answers)
```

# Question time series

### Question volume

```python
treated = questions_T_resampled[['PostTypeId']].count()
control = questions_C_resampled[['PostTypeId']].count()
so.plot_ts(treated, control, 'PostTypeId', 'Question Volume', 'qa/q.pdf')
```

### Question views

#### Mean

```python
treated = questions_T_resampled[['ViewCount']].mean()
control = questions_C_resampled[['ViewCount']].mean()
so.plot_ts(treated, control, 'ViewCount', 'Question Views', 'qa/q_views.pdf', 'Mean views')
```

#### Median

```python
treated = questions_T_resampled[['ViewCount']].median()
control = questions_C_resampled[['ViewCount']].median()
so.plot_ts(treated, control, 'ViewCount', 'Question Views', 'qa/median/q_views.pdf', 'Median views')
```
